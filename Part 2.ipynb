```python
# Q9
##### Plotting experiments
epochs = []
accuracies = []
losses = []

e = []
a = []
l = []

with open('results/mnist_relu.txt', 'r') as infile:
    contents = infile.read()
contents = contents.split('\n')
contents = [x.strip(' ') for x in contents]
for i in range(0, len(contents), 3):
    epochs.append(int(contents[i].split(' ')[1]))
    accuracies.append(float(contents[i+1].split(' ')[1]))
    losses.append(float(contents[i+2].split(' ')[2]))

with open('results/mnist_sigm.txt', 'r') as infile:
    contents = infile.read()
contents = contents.split('\n')
contents = [x.strip(' ') for x in contents]
for i in range(0, len(contents), 3):
    e.append(int(contents[i].split(' ')[1]))
    a.append(float(contents[i+1].split(' ')[1]))
    l.append(float(contents[i+2].split(' ')[2]))
    

figure, axis = plt.subplots(2, 1, dpi=300, figsize=(5, 3), sharex=True)


plt.xticks(epochs, [str(e+1) for e in epochs])
axis[0].plot(epochs, accuracies, c='black', linewidth=0.8)
axis[1].plot(e, a, c='black', linewidth=0.8);

axis[0].set_title('RELU', y=0.93, size=8)
axis[1].set_title('SIGMOID', y=0.93, size=8)
plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy', y=1.2)


plt.show();
```
