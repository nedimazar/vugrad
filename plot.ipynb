```python
# Q9
##### Plotting experiments
def getit(path):
    e = []
    a = []
    l = []
    with open(path, 'r') as infile:
        contents = infile.read()
    contents = contents.split('\n')
    contents = [x.strip(' ') for x in contents]
    for i in range(0, len(contents), 3):
        e.append(int(contents[i].split(' ')[1]))
        a.append(float(contents[i+1].split(' ')[1]))
        l.append(float(contents[i+2].split(' ')[2]))
    return e, a, l
    
epochs, accuracies, losses = getit('results/vu_model.txt')

e, a, l = getit('results/hidden8x.txt')

ep, ac, lo = getit('results/hidden2x.txt')

figure, axis = plt.subplots(3, 1, dpi=300, figsize=(5, 3), sharex=True)
# axis[0].grid()
# axis[1].grid()
plt.ylim([0, 1])
plt.xticks(epochs, [str(e+1) for e in epochs])
axis[0].plot(ep, ac, c='blue', linewidth=0.8)
axis[0].plot(ep, lo, c='red', linewidth=0.8)

axis[1].plot(epochs, accuracies, c='blue', linewidth=0.8);
axis[1].plot(epochs, losses, c='red', linewidth=0.8);

axis[2].plot(e, a, c='blue', linewidth=0.8);
axis[2].plot(e, l, c='red', linewidth=0.8);


axis[0].set_title('Hidden 2x', y=0.93, size=8)
axis[1].set_title('Hidden 4x', y=0.93, size=8)
axis[2].set_title('Hidden 8x', y=0.93, size=8)



axis[0].legend(['Accuracy', 'Loss'])
plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy', y=1.2)

plt.show();
```
